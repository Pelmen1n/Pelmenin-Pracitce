# Задание 1

Симулятор виртуального питомца

Цель: создать виртуальный симулятор домашних животных, в котором пользователи смогут заводить питомцев и ухаживать за ними.

Требования:

- Функция для усыновления питомца (имя, тип, возраст).
- Функция для того, чтобы покормить питомца, поиграть с ним или уложить его спать.
- Функция для отображения состояния питомца (голод, радость, энергия).

````python
import random

def answers():
    print('Добропожаловать в питомник, здесь вы можете найти любого питомца')
    print('Введите харакетристики Питомца')
    name = input('Введите имя питомца: ')
    age = int(input('Введите возраст питомца: '))
    while age >= 10:
        print('Таких старых питомцев у нас нету, давайте выберем другой возраст(Меньше 10)')
        age = int(input('Введите возраст питомца: '))
    animal = input('Введите тип животного (Собака, Кошка, Морская свинка, крот): ')
    info = []
    info.append(name)
    info.append(age)
    info.append(animal)
    print('Отлично, теперь пожалуйста позаботьтесь о своем питомце')
    print('Вы можете накормить его, положить спать, поиграть и посмотреть его состояние')
    return f"Имя - {info[0]}, Возраст - {info[1]}, Тип - {info[2]}"


def randomize(stats):
    print('-----------------------')
    print('Сейчас у вашего питомца такие показатели :')
    hunger = random.randint(5, 35)
    happines = random.randint(0, 50)
    energy = random.randint(15, 40)
    stats.append(hunger)
    stats.append(happines)
    stats.append(energy)
    print(f"Голод - {stats[0]}%, Радость - {stats[1]}%, Энергия - {stats[2]}%")
    return stats

def tasks():
    stats = []
    randomize(stats)
    while stats[0] + stats[1] + stats[2] < 300:
        print('<1> Покормить; <2> Поиграть; <3> Уложить спать; <4> Показатели Питомца; <5> Я передумал')
        question = input('Что вы хотите сделать с питомцем?  ')
        if question == "1":
            feed = input('<1> Наполнить кормушку; <2> Наполнить поилку  ')
            if feed == '1':
                stats[0] += 25
                if stats[0] > 100:
                    stats[0] = 100
                    print(f'Питомец наелся')
                print(f'Теперь голод вашего питомца составляет - {stats[0]}%')
            elif feed == '2':
                stats[0] += 15
                if stats[0] > 100:
                    stats[0] = 100
                    print(f'Питомец наелся')
                print(f'Теперь голод вашего питомца составляет - {stats[0]}%')
            else:
                print('Неправильный ввод')
            print('-----------------------')
        elif question == "2":
            if stats[2] > 10:
                stats[1] += 35
                stats[2] -= 10
                if stats[1] > 100:
                    stats[1] = 100
                print(f'Питомец счастлив')
                print(f'Теперь радость вашего питомца составляет - {stats[1]}%, а энергия - {stats[2]}%')
            else:
                print('Вы не можете играть, ваш питомец хочет спать')
            print('-----------------------')
        elif question == "3":
            if stats[0] > 25:
                stats[2] += 35
                stats[0] -= 10
                if stats[2] > 100:
                    stats[2] = 100
                    print(f'Питомец выспался')
                print(f'Теперь энергия вашего питомца составляет - {stats[2]}%, а голод - {stats[0]}%')
            else:
                print('Покормите ваше чадо, прежде чем идти спать')
            print('-----------------------')
        elif question == "4":
            print(f"Голод - {stats[0]}%, Радость - {stats[1]}%, Энергия - {stats[2]}%")
            print('-----------------------')
        elif question == "5":
            break
    else:
        print('Вы крутой, ваш питомец имеет фулл статы!')


print(answers())

tasks()
````

# Задание 2

Рыцарь и дракон

Цель: создать небольшую игру, в которой вам необходимо играть за рыцаря и сразиться с драконом

Требования:

- Создание персонажа (имя, информация о доспехах, оружии, урон, здоровье)
- Управление персонажем и мини сюжет
- Создание дракона (Имя, информация о здоровье и уроне)
- Боевая система (нанесение и получение урона, урон должен быть случайным в заданном диапазоне)
- Реализовать бой между драконом и рыцарем

````python
import random

# В этот раз без артемия лебедева (

def create_character(character_type):
    name = input(f"Введите имя вашего {character_type}: ")
    health = int(input(f"Введите здоровье вашего {character_type}: "))
    if character_type == "рыцаря":
        weapon = input(f"Выберите оружие для вашего {character_type}: ")
    else:
        weapon = "артеми лебеде" # Тссс 🤫
    base_damage = int(input(f"Введите урон вашего {character_type}: "))
    return {"name": name, "health": health, "weapon": weapon, "base_damage": base_damage}


def attack(attacker, defender):
    damage = random.randint(max(1, attacker["base_damage"] - 10), attacker["base_damage"] + 10)
    defender["health"] -= damage
    return damage


def block(character, damage):
    blocked_damage = damage // 2
    character["health"] -= blocked_damage
    return blocked_damage


def dragon_attack(dragon):
    if random.choice([True, False]):
        print("Дракон готовится к критическому удару!")
        return random.randint(dragon["base_damage"] + 10, dragon["base_damage"] + 20)
    else:
        return random.randint(max(1, dragon["base_damage"] - 10), dragon["base_damage"] + 10)


print("Добро пожаловать в игру!")
print(
    "Ты - рыцарь, которому нечего делать, по этому ты решил найти себе дракона чтобы подраться с ним")
print("Тебе предстоит сразиться с драконом, которого ты создашь сам. Удачи :)")

knight = create_character("рыцаря")

dragon = create_character("дракона")

while knight["health"] > 0 and dragon["health"] > 0:
    print(f"\n{knight['name']} (Здоровье: {knight['health']}) vs {dragon['name']} (Здоровье: {dragon['health']})")

    dragon_damage = dragon_attack(dragon)

    action = input("Выберите действие (атака/блок): ").lower()

    if action == "атака":
        damage = attack(knight, dragon)
        print(f"{knight['name']} атакует {dragon['name']} и наносит {damage} урона")
    elif action == "блок":
        blocked_damage = block(knight, dragon_damage)
        print(f"{knight['name']} блокирует атаку и получает {blocked_damage} урона")
    else:
        print("Неправильный ввод")
        continue

    if dragon["health"] <= 0:
        print(f"{dragon['name']} побежден, ты крутой")
        break

    if action != "блок":
        knight["health"] -= dragon_damage
        print(f"{dragon['name']} атакует и наносит {dragon_damage} урона {knight['name']}")

    if knight["health"] <= 0:
        print(f"{knight['name']} побежден, анлак :(")
        break
````
