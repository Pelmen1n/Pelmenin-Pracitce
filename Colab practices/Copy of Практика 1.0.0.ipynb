# Задание 1

Задание:

Создайте класс с именем Rectangle который имеет:
- Атрибуты ширины и высоты.
- Метод расчета площади.
- Метод расчета периметра.
- Метод отображения размеров прямоугольника.

Создайте экземпляр класса Rectangleи продемонстрируйте его функциональность.

```python
class Rectangle:

    def __init__(self, height, width):
        self.height = height
        self.width = width

    def rectangle_area(self):
        area = self.height * self.width
        return area

    def rectangle_perimetr(self):
        perimetr = (self.height + self.width) * 2
        return perimetr

    def rectangle_display(self):
        return f"Высота - {self.height}; Длина - {self.width}"

rect1 = Rectangle(3, 8)

print(rect1.rectangle_display())
print(f"Площадь : {rect1.rectangle_area()}")
print(f"Периметр : {rect1.rectangle_perimetr()}")

```

# Задание 2

Задание: Создайте мини версию банковской системы:


Инструкции:

1. Создайте класс BankAccountсо следующими атрибутами:
    - account_holder -  владелец счета
    - balance - баланс счета

2. Реализуйте следующие методы:
    - Метод для инициализации владельца счета: имя владельца счета и установите начальный баланс на 0.
    - deposit(amount): Добавьте указанную сумму к балансу.
    - withdraw(amount): Вычесть указанную сумму из баланса, если средств достаточно; в противном случае вывести предупреждение.
    - get_balance(): Возврат текущего баланса.


Создайте объект класса и продемонстрируйте его возможности

```python
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def initialization(self):
        self.balance = 0

    def deposit(self, amount):
        amount = float(input("Введите сумму пополнения: "))
        self.balance += amount
        print(f"Счет пополнен на {amount}₽. Текущий баланс: {self.balance}₽")

    def withdraw(self, amount):
        amount = float(input("Введите сумму снятия: "))
        if amount > self.balance:
            print(f"На вашем аккаунте недостаточно средств. Текущий баланс: {self.balance}₽")
        else:
            self.balance -= amount
            print(f"Снято {amount}₽. Текущий баланс: {self.balance}₽")

    def get_balance(self):
        return f"На счете {self.account_holder} лежит {self.balance}₽"

print("Здравствуйте, это Z-Bank. Для создания счета заполните анкету")
account_holder = input("Придумайте имя владельца счета: ")
account = BankAccount(account_holder)

while True:
    print("\n1.Пополнить счет 2.Снять деньги 3.Показать баланс 4.Выход")
    answer = input()
    if answer == '1':
        account.deposit(0)
    elif answer == '2':
        account.withdraw(0)
    elif answer == '3':
        print(account.get_balance())
    elif answer == '4':
        print("Мне пора, удачи!")
        break
    else:
        print("Неверный ввод, попробуйте еще раз")
```
