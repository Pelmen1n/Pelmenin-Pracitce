# –ó–∞–¥–∞–Ω–∏–µ 1

–ó–∞–¥–∞–Ω–∏–µ:

–°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å —Å –∏–º–µ–Ω–µ–º Rectangle –∫–æ—Ç–æ—Ä—ã–π –∏–º–µ–µ—Ç:
- –ê—Ç—Ä–∏–±—É—Ç—ã —à–∏—Ä–∏–Ω—ã –∏ –≤—ã—Å–æ—Ç—ã.
- –ú–µ—Ç–æ–¥ —Ä–∞—Å—á–µ—Ç–∞ –ø–ª–æ—â–∞–¥–∏.
- –ú–µ—Ç–æ–¥ —Ä–∞—Å—á–µ—Ç–∞ –ø–µ—Ä–∏–º–µ—Ç—Ä–∞.
- –ú–µ—Ç–æ–¥ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–æ–≤ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞.

–°–æ–∑–¥–∞–π—Ç–µ —ç–∫–∑–µ–º–ø–ª—è—Ä –∫–ª–∞—Å—Å–∞ Rectangle–∏ –ø—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ –µ–≥–æ —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å.

```python
class Rectangle:

    def __init__(self, height, width):
        self.height = height
        self.width = width

    def rectangle_area(self):
        area = self.height * self.width
        return area

    def rectangle_perimetr(self):
        perimetr = (self.height + self.width) * 2
        return perimetr

    def rectangle_display(self):
        return f"–í—ã—Å–æ—Ç–∞ - {self.height}; –î–ª–∏–Ω–∞ - {self.width}"

rect1 = Rectangle(3, 8)

print(rect1.rectangle_display())
print(f"–ü–ª–æ—â–∞–¥—å : {rect1.rectangle_area()}")
print(f"–ü–µ—Ä–∏–º–µ—Ç—Ä : {rect1.rectangle_perimetr()}")

```

# –ó–∞–¥–∞–Ω–∏–µ 2

–ó–∞–¥–∞–Ω–∏–µ: –°–æ–∑–¥–∞–π—Ç–µ –º–∏–Ω–∏ –≤–µ—Ä—Å–∏—é –±–∞–Ω–∫–æ–≤—Å–∫–æ–π —Å–∏—Å—Ç–µ–º—ã:


–ò–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏:

1. –°–æ–∑–¥–∞–π—Ç–µ –∫–ª–∞—Å—Å BankAccount—Å–æ —Å–ª–µ–¥—É—é—â–∏–º–∏ –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏:
    - account_holder -  –≤–ª–∞–¥–µ–ª–µ—Ü —Å—á–µ—Ç–∞
    - balance - –±–∞–ª–∞–Ω—Å —Å—á–µ—Ç–∞

2. –†–µ–∞–ª–∏–∑—É–π—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ –º–µ—Ç–æ–¥—ã:
    - –ú–µ—Ç–æ–¥ –¥–ª—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –≤–ª–∞–¥–µ–ª—å—Ü–∞ —Å—á–µ—Ç–∞: –∏–º—è –≤–ª–∞–¥–µ–ª—å—Ü–∞ —Å—á–µ—Ç–∞ –∏ —É—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –Ω–∞—á–∞–ª—å–Ω—ã–π –±–∞–ª–∞–Ω—Å –Ω–∞ 0.
    - deposit(amount): –î–æ–±–∞–≤—å—Ç–µ —É–∫–∞–∑–∞–Ω–Ω—É—é —Å—É–º–º—É –∫ –±–∞–ª–∞–Ω—Å—É.
    - withdraw(amount): –í—ã—á–µ—Å—Ç—å —É–∫–∞–∑–∞–Ω–Ω—É—é —Å—É–º–º—É –∏–∑ –±–∞–ª–∞–Ω—Å–∞, –µ—Å–ª–∏ —Å—Ä–µ–¥—Å—Ç–≤ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ; –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–µ—Å—Ç–∏ –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ.
    - get_balance(): –í–æ–∑–≤—Ä–∞—Ç —Ç–µ–∫—É—â–µ–≥–æ –±–∞–ª–∞–Ω—Å–∞.


–°–æ–∑–¥–∞–π—Ç–µ –æ–±—ä–µ–∫—Ç –∫–ª–∞—Å—Å–∞ –∏ –ø—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–π—Ç–µ –µ–≥–æ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏

```python
class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def initialization(self):
        self.balance = 0

    def deposit(self, amount):
        amount = float(input("–í–≤–µ–¥–∏—Ç–µ —Å—É–º–º—É –ø–æ–ø–æ–ª–Ω–µ–Ω–∏—è: "))
        self.balance += amount
        print(f"–°—á–µ—Ç –ø–æ–ø–æ–ª–Ω–µ–Ω –Ω–∞ {amount}‚ÇΩ. –¢–µ–∫—É—â–∏–π –±–∞–ª–∞–Ω—Å: {self.balance}‚ÇΩ")

    def withdraw(self, amount):
        amount = float(input("–í–≤–µ–¥–∏—Ç–µ —Å—É–º–º—É —Å–Ω—è—Ç–∏—è: "))
        if amount > self.balance:
            print(f"–ù–∞ –≤–∞—à–µ–º –∞–∫–∫–∞—É–Ω—Ç–µ –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Å—Ä–µ–¥—Å—Ç–≤. –¢–µ–∫—É—â–∏–π –±–∞–ª–∞–Ω—Å: {self.balance}‚ÇΩ")
        else:
            self.balance -= amount
            print(f"–°–Ω—è—Ç–æ {amount}‚ÇΩ. –¢–µ–∫—É—â–∏–π –±–∞–ª–∞–Ω—Å: {self.balance}‚ÇΩ")

    def get_balance(self):
        return f"–ù–∞ —Å—á–µ—Ç–µ {self.account_holder} –ª–µ–∂–∏—Ç {self.balance}‚ÇΩ"

print("–ó–¥—Ä–∞–≤—Å—Ç–≤—É–π—Ç–µ, —ç—Ç–æ Z-Bank. –î–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Å—á–µ—Ç–∞ –∑–∞–ø–æ–ª–Ω–∏—Ç–µ –∞–Ω–∫–µ—Ç—É")
account_holder = input("–ü—Ä–∏–¥—É–º–∞–π—Ç–µ –∏–º—è –≤–ª–∞–¥–µ–ª—å—Ü–∞ —Å—á–µ—Ç–∞: ")
account = BankAccount(account_holder)

while True:
    print("\n1.–ü–æ–ø–æ–ª–Ω–∏—Ç—å —Å—á–µ—Ç 2.–°–Ω—è—Ç—å –¥–µ–Ω—å–≥–∏ 3.–ü–æ–∫–∞–∑–∞—Ç—å –±–∞–ª–∞–Ω—Å 4.–í—ã—Ö–æ–¥")
    answer = input()
    if answer == '1':
        account.deposit(0)
    elif answer == '2':
        account.withdraw(0)
    elif answer == '3':
        print(account.get_balance())
    elif answer == '4':
        print("–ú–Ω–µ –ø–æ—Ä–∞, —É–¥–∞—á–∏!")
        break
    else:
        print("–ù–µ–≤–µ—Ä–Ω—ã–π –≤–≤–æ–¥, –ø–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑")
```

# –ó–∞–¥–∞–Ω–∏–µ 3

–í–æ–∑—å–º–∏—Ç–µ –∫–æ–¥ –∏ –∑–∞–¥–∞–Ω–∏–µ (–†—ã—Ü–∞—Ä—å –∏ –¥—Ä–∞–∫–æ–Ω) –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –ø—Ä–∞–∫—Ç–∏–∫–∏ –∏ —Ä–µ–∞–ª–∏–∑—É–π—Ç–µ –µ–≥–æ —Å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º –∫–ª–∞—Å—Å–æ–≤

```python
import random

# –≤—Å–µ, –≤—Å–µ, –≤—Å–µ –±–æ–ª—å—à–µ –Ω–∏–∫–∞–∫–∏—Ö —É—Ç–∫–∏–Ω—ã—Ö –∏ –ª–µ–±–µ–¥–µ–≤—ã—Ö

class Character:
    def __init__(self, name, health, base_damage):
        self.name = name
        self.health = health
        self.base_damage = base_damage

    def attack(self, other):
        damage = random.randint(max(1, self.base_damage - 10), self.base_damage + 10)
        other.health -= damage
        return damage

    def block(self, damage):
        blocked_damage = damage // 2
        self.health -= blocked_damage
        return blocked_damage


def create_character(character_type):
    name = input(f"–í–≤–µ–¥–∏—Ç–µ –∏–º—è –≤–∞—à–µ–≥–æ {character_type}: ")
    health = int(input(f"–í–≤–µ–¥–∏—Ç–µ –∑–¥–æ—Ä–æ–≤—å–µ –≤–∞—à–µ–≥–æ {character_type}: "))
    if character_type == "—Ä—ã—Ü–∞—Ä—è":
        weapon = input(f"–í—ã–±–µ—Ä–∏—Ç–µ –æ—Ä—É–∂–∏–µ –¥–ª—è –≤–∞—à–µ–≥–æ {character_type}: ")
    else:
        weapon = "–æ–≥–æ–Ω—å"
    base_damage = int(input(f"–í–≤–µ–¥–∏—Ç–µ —É—Ä–æ–Ω –≤–∞—à–µ–≥–æ {character_type}: "))
    return Character(name, health, base_damage)


print("–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ –∏–≥—Ä—É!")
print(
    "–¢—ã - —Ä—ã—Ü–∞—Ä—å, –∫–æ—Ç–æ—Ä–æ–º—É –Ω–µ—á–µ–≥–æ –¥–µ–ª–∞—Ç—å, –ø–æ —ç—Ç–æ–º—É —Ç—ã —Ä–µ—à–∏–ª –Ω–∞–π—Ç–∏ —Å–µ–±–µ –¥—Ä–∞–∫–æ–Ω–∞ —á—Ç–æ–±—ã –ø–æ–¥—Ä–∞—Ç—å—Å—è —Å –Ω–∏–º")
print("–¢–µ–±–µ –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç —Å—Ä–∞–∑–∏—Ç—å—Å—è —Å –¥—Ä–∞–∫–æ–Ω–æ–º, –∫–æ—Ç–æ—Ä–æ–≥–æ —Ç—ã —Å–æ–∑–¥–∞—à—å —Å–∞–º. –£–¥–∞—á–∏ :)")

knight = create_character("—Ä—ã—Ü–∞—Ä—è")

dragon = create_character("–¥—Ä–∞–∫–æ–Ω–∞")

while knight.health > 0 and dragon.health > 0:
    print(f"\n{knight.name} (–ó–¥–æ—Ä–æ–≤—å–µ: {knight.health}) vs {dragon.name} (–ó–¥–æ—Ä–æ–≤—å–µ: {dragon.health})")

    dragon_damage = random.randint(max(1, dragon.base_damage - 10), dragon.base_damage + 10)
    if random.choice([True, False]):
        print("–î—Ä–∞–∫–æ–Ω –≥–æ—Ç–æ–≤–∏—Ç—Å—è –∫ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–º—É —É–¥–∞—Ä—É!")
        dragon_damage = random.randint(dragon.base_damage + 10, dragon.base_damage + 20)

    action = input("–í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ (–∞—Ç–∞–∫–∞/–±–ª–æ–∫): ").lower()

    if action == "–∞—Ç–∞–∫–∞":
        damage = knight.attack(dragon)
        print(f"{knight.name} –∞—Ç–∞–∫—É–µ—Ç {dragon.name} –∏ –Ω–∞–Ω–æ—Å–∏—Ç {damage} —É—Ä–æ–Ω–∞.")
    elif action == "–±–ª–æ–∫":
        blocked_damage = knight.block(dragon_damage)
        print(f"{knight.name} –±–ª–æ–∫–∏—Ä—É–µ—Ç –∞—Ç–∞–∫—É –∏ –ø–æ–ª—É—á–∞–µ—Ç {blocked_damage} —É—Ä–æ–Ω–∞.")
    elif action == "–¥–æ–¥–æ –ø–∏—Ü—Ü–∞":
        print(f"–ù—É –Ω–µ, –Ω—É –ø–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω—É –Ω–µ–µ–µ—Ç http://surl.li/eeyosh")  #–ß—Ç–æ –±—ã —ç—Ç–æ –º–æ–≥–ª–æ –∑–Ω–∞—á–∏—Ç—åüßê
    else:
        print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤–≤–æ–¥")
        continue

    if dragon.health <= 0:
        print(f"{dragon.name} –ø–æ–±–µ–∂–¥–µ–Ω, —Ç—ã –∫—Ä—É—Ç–æ–π")
        break

    if action != "–±–ª–æ–∫":
        knight.health -= dragon_damage
        print(f"{dragon.name} –∞—Ç–∞–∫—É–µ—Ç –∏ –Ω–∞–Ω–æ—Å–∏—Ç {dragon_damage} —É—Ä–æ–Ω–∞ {knight.name}.")

    if knight.health <= 0:
        print(f"{knight.name} –ø–æ–±–µ–∂–¥–µ–Ω, –∞–Ω–ª–∞–∫ :(")
        break
```
