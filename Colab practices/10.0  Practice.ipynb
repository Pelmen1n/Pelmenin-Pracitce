# Задание 1

**Описание:** Создайте иерархию классов для разных типов сотрудников в компании. Реализуйте родительский класс Employee и дочерние классы Manager и Developer. Каждый класс должен иметь метод для расчета зарплаты на основе различных критериев класса.


Отрабатываемый принцип: Наследование



```python
class Employee:

    def __init__(self, name, salary, job_title):
        self.name = name
        self.salary = salary
        self.job_title = job_title

    def base_salary(self):
        return self.salary

    def __str__(self):
        return f"Сотрудник {self.name}, должности - {self.job_title} получает {self.base_salary()}$"

class Manager(Employee):

    def __init__(self, name, salary, job_title, job_rate):
        super().__init__(name, salary, job_title)
        self.job_rate = job_rate

    def base_salary(self):
        return self.salary * self.job_rate

class Developer(Employee):

    def __init__(self, name, salary, job_title, overtime_hours, overtime_rate):
        super().__init__(name, salary, job_title)
        self.overtime_hours = overtime_hours
        self.overtime_rate = overtime_rate

    def base_salary(self):
        return self.salary + self.overtime_hours * self.overtime_rate

employee = Employee("Игорь", 300, "Стажер")
manager = Manager("Олег", 700, "Менеджер по продажам", 1.5)
developer = Developer("Макс", 600, "Программист", 12, 75)

proffessions = [employee, manager, developer]
for proffession in proffessions:
    print(proffession)
```

# Задание 2

**Описание:** Создайте иерархию классов для различных типов транспортных средств (Необходим один родительский класс и 3 дочерних). Реализуйте метод, который позволяет каждому транспортному средству возвращать собственное описание (Метод в каждом классе должен иметь одинаковое название). Продемонстрируйте вызов данного метода для каждого транспортного средства.


Отрабатываемый принцип: Полиморфизм

```python
class Vehicle:

    def __init__(self, type, speed):
        self.type = type
        self.speed = speed

    def description(self):
        return f"{self.type} движется со скоростью {self.speed} км/ч"

class Car(Vehicle):

    def __init__(self, name, speed, fuel_type):
        super().__init__(name, speed)
        self.fuel_type = fuel_type

    def description(self):
        return f"Автомобиль {self.type} едет с максимальной скоростью {self.speed} км/ч, тип топлива: {self.fuel_type}"

class Plane(Vehicle):
    def __init__(self, name, speed, type_plane):
        super().__init__(name, speed)
        self.type_plane = type_plane

    def description(self):
        return f"Самолет {self.type} летит со скоростью {self.speed} км/ч, тип: {self.type_plane}"

class Boat(Vehicle):
    def __init__(self, name, speed, capacity):
        super().__init__(name, speed)
        self.capacity = capacity

    def description(self):
        return f"Судно {self.type} движется со скоростью {self.speed} км/ч, вместимость: {self.capacity} человек"



car = Car("Tesla Model S", 250, "электричество")
plane = Plane("Boeing 777", 920, "пассажирский")
boat = Boat("Яхта", 50, 10)


vehicles = [car, plane, boat]
for vehicle in vehicles:
    print(vehicle.description())
```
