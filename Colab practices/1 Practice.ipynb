Задание 7: Необходимо создать простой калькулятор, который позволяет пользователю выбрать одну из четырёх операций (+, -, *, /), ввести два числа и получить результат выполнения операции.

`Введите операцию:`

`Введите первое число:`

`Введите второе число:`

`Результат выполнения: `


```python
oper = str(input())
num1 = int(input())
num2 = int(input())
if oper == '+':
  print(num1+num2)
elif oper == '-':
  print(num1-num2)
elif oper == '*':
  print(num1*num2)
else:
  print(num1/num2)
```

# Практика 0.2.1

**Шахматы**

Даны стартовые и конечные координаты, а также фигура

Необходимо определить, может ли заданная фигура так ходить?

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())
figura = str(input())
if figura == 'пешка':
  if y2-y1 <= 1:
    print('можно')
  else:
    print('ne')
elif figura == 'ладья':
  if x1==x2 or y1==y2:
    print('можно')
  else:
    print('ne')
elif figura == 'слон':
  if y1-x1 == y2-x2 or abs(y1+x1) == abs(y2+x2):
    print('можно')
  else:
    print('ne')
elif figura == 'конь':
  if abs(x1-x2) == 2 and abs(y1-y2) == 1 or abs(x1-x2) == 1 and abs(y1-y2) == 2:
    print('можно')
  else:
    print('ne')
elif figura == 'король':
  if (x1 == x2-1 or x1 == x2+1 or x1 == x2) and (y1 == y2-1 or y1 == y2+1 or y1 == y2):
    print('можно')
  else:
    print('ne')
elif figura == 'ферзь':
  if y1-x1 == y2-x2 or abs(y1+x1) == abs(y2+x2):
    print('можно')
  elif x1 == x2 or y1 == y2:
    print('можно')
  else:
    print('ne')
else:
  print('вы ввели неправильное название фигуры')

```
