# –ó–∞–¥–∞–Ω–∏–µ 1

–°–∏–º—É–ª—è—Ç–æ—Ä –≤–∏—Ä—Ç—É–∞–ª—å–Ω–æ–≥–æ –ø–∏—Ç–æ–º—Ü–∞

–¶–µ–ª—å: —Å–æ–∑–¥–∞—Ç—å –≤–∏—Ä—Ç—É–∞–ª—å–Ω—ã–π —Å–∏–º—É–ª—è—Ç–æ—Ä –¥–æ–º–∞—à–Ω–∏—Ö –∂–∏–≤–æ—Ç–Ω—ã—Ö, –≤ –∫–æ—Ç–æ—Ä–æ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å–º–æ–≥—É—Ç –∑–∞–≤–æ–¥–∏—Ç—å –ø–∏—Ç–æ–º—Ü–µ–≤ –∏ —É—Ö–∞–∂–∏–≤–∞—Ç—å –∑–∞ –Ω–∏–º–∏.

–¢—Ä–µ–±–æ–≤–∞–Ω–∏—è:

- –§—É–Ω–∫—Ü–∏—è –¥–ª—è —É—Å—ã–Ω–æ–≤–ª–µ–Ω–∏—è –ø–∏—Ç–æ–º—Ü–∞ (–∏–º—è, —Ç–∏–ø, –≤–æ–∑—Ä–∞—Å—Ç).
- –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –ø–æ–∫–æ—Ä–º–∏—Ç—å –ø–∏—Ç–æ–º—Ü–∞, –ø–æ–∏–≥—Ä–∞—Ç—å —Å –Ω–∏–º –∏–ª–∏ —É–ª–æ–∂–∏—Ç—å –µ–≥–æ —Å–ø–∞—Ç—å.
- –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Å–æ—Å—Ç–æ—è–Ω–∏—è –ø–∏—Ç–æ–º—Ü–∞ (–≥–æ–ª–æ–¥, —Ä–∞–¥–æ—Å—Ç—å, —ç–Ω–µ—Ä–≥–∏—è).

````python
import random

def answers():
    print('–î–æ–±—Ä–æ–ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ –ø–∏—Ç–æ–º–Ω–∏–∫, –∑–¥–µ—Å—å –≤—ã –º–æ–∂–µ—Ç–µ –Ω–∞–π—Ç–∏ –ª—é–±–æ–≥–æ –ø–∏—Ç–æ–º—Ü–∞')
    print('–í–≤–µ–¥–∏—Ç–µ —Ö–∞—Ä–∞–∫–µ—Ç—Ä–∏—Å—Ç–∏–∫–∏ –ü–∏—Ç–æ–º—Ü–∞')
    name = input('–í–≤–µ–¥–∏—Ç–µ –∏–º—è –ø–∏—Ç–æ–º—Ü–∞: ')
    age = int(input('–í–≤–µ–¥–∏—Ç–µ –≤–æ–∑—Ä–∞—Å—Ç –ø–∏—Ç–æ–º—Ü–∞: '))
    while age >= 10:
        print('–¢–∞–∫–∏—Ö —Å—Ç–∞—Ä—ã—Ö –ø–∏—Ç–æ–º—Ü–µ–≤ —É –Ω–∞—Å –Ω–µ—Ç—É, –¥–∞–≤–∞–π—Ç–µ –≤—ã–±–µ—Ä–µ–º –¥—Ä—É–≥–æ–π –≤–æ–∑—Ä–∞—Å—Ç(–ú–µ–Ω—å—à–µ 10)')
        age = int(input('–í–≤–µ–¥–∏—Ç–µ –≤–æ–∑—Ä–∞—Å—Ç –ø–∏—Ç–æ–º—Ü–∞: '))
    animal = input('–í–≤–µ–¥–∏—Ç–µ —Ç–∏–ø –∂–∏–≤–æ—Ç–Ω–æ–≥–æ (–°–æ–±–∞–∫–∞, –ö–æ—à–∫–∞, –ú–æ—Ä—Å–∫–∞—è —Å–≤–∏–Ω–∫–∞, –∫—Ä–æ—Ç): ')
    info = []
    info.append(name)
    info.append(age)
    info.append(animal)
    print('–û—Ç–ª–∏—á–Ω–æ, —Ç–µ–ø–µ—Ä—å –ø–æ–∂–∞–ª—É–π—Å—Ç–∞ –ø–æ–∑–∞–±–æ—Ç—å—Ç–µ—Å—å –æ —Å–≤–æ–µ–º –ø–∏—Ç–æ–º—Ü–µ')
    print('–í—ã –º–æ–∂–µ—Ç–µ –Ω–∞–∫–æ—Ä–º–∏—Ç—å –µ–≥–æ, –ø–æ–ª–æ–∂–∏—Ç—å —Å–ø–∞—Ç—å, –ø–æ–∏–≥—Ä–∞—Ç—å –∏ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –µ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏–µ')
    return f"–ò–º—è - {info[0]}, –í–æ–∑—Ä–∞—Å—Ç - {info[1]}, –¢–∏–ø - {info[2]}"


def randomize(stats):
    print('-----------------------')
    print('–°–µ–π—á–∞—Å —É –≤–∞—à–µ–≥–æ –ø–∏—Ç–æ–º—Ü–∞ —Ç–∞–∫–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ :')
    hunger = random.randint(5, 35)
    happines = random.randint(0, 50)
    energy = random.randint(15, 40)
    stats.append(hunger)
    stats.append(happines)
    stats.append(energy)
    print(f"–ì–æ–ª–æ–¥ - {stats[0]}%, –†–∞–¥–æ—Å—Ç—å - {stats[1]}%, –≠–Ω–µ—Ä–≥–∏—è - {stats[2]}%")
    return stats

def tasks():
    stats = []
    randomize(stats)
    while stats[0] + stats[1] + stats[2] < 300:
        print('<1> –ü–æ–∫–æ—Ä–º–∏—Ç—å; <2> –ü–æ–∏–≥—Ä–∞—Ç—å; <3> –£–ª–æ–∂–∏—Ç—å —Å–ø–∞—Ç—å; <4> –ü–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ü–∏—Ç–æ–º—Ü–∞; <5> –Ø –ø–µ—Ä–µ–¥—É–º–∞–ª')
        question = input('–ß—Ç–æ –≤—ã —Ö–æ—Ç–∏—Ç–µ —Å–¥–µ–ª–∞—Ç—å —Å –ø–∏—Ç–æ–º—Ü–µ–º?  ')
        if question == "1":
            feed = input('<1> –ù–∞–ø–æ–ª–Ω–∏—Ç—å –∫–æ—Ä–º—É—à–∫—É; <2> –ù–∞–ø–æ–ª–Ω–∏—Ç—å –ø–æ–∏–ª–∫—É  ')
            if feed == '1':
                stats[0] += 25
                if stats[0] > 100:
                    stats[0] = 100
                    print(f'–ü–∏—Ç–æ–º–µ—Ü –Ω–∞–µ–ª—Å—è')
                print(f'–¢–µ–ø–µ—Ä—å –≥–æ–ª–æ–¥ –≤–∞—à–µ–≥–æ –ø–∏—Ç–æ–º—Ü–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç - {stats[0]}%')
            elif feed == '2':
                stats[0] += 15
                if stats[0] > 100:
                    stats[0] = 100
                    print(f'–ü–∏—Ç–æ–º–µ—Ü –Ω–∞–µ–ª—Å—è')
                print(f'–¢–µ–ø–µ—Ä—å –≥–æ–ª–æ–¥ –≤–∞—à–µ–≥–æ –ø–∏—Ç–æ–º—Ü–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç - {stats[0]}%')
            else:
                print('–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤–≤–æ–¥')
            print('-----------------------')
        elif question == "2":
            if stats[2] > 10:
                stats[1] += 35
                stats[2] -= 10
                if stats[1] > 100:
                    stats[1] = 100
                print(f'–ü–∏—Ç–æ–º–µ—Ü —Å—á–∞—Å—Ç–ª–∏–≤')
                print(f'–¢–µ–ø–µ—Ä—å —Ä–∞–¥–æ—Å—Ç—å –≤–∞—à–µ–≥–æ –ø–∏—Ç–æ–º—Ü–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç - {stats[1]}%, –∞ —ç–Ω–µ—Ä–≥–∏—è - {stats[2]}%')
            else:
                print('–í—ã –Ω–µ –º–æ–∂–µ—Ç–µ –∏–≥—Ä–∞—Ç—å, –≤–∞—à –ø–∏—Ç–æ–º–µ—Ü —Ö–æ—á–µ—Ç —Å–ø–∞—Ç—å')
            print('-----------------------')
        elif question == "3":
            if stats[0] > 25:
                stats[2] += 35
                stats[0] -= 10
                if stats[2] > 100:
                    stats[2] = 100
                    print(f'–ü–∏—Ç–æ–º–µ—Ü –≤—ã—Å–ø–∞–ª—Å—è')
                print(f'–¢–µ–ø–µ—Ä—å —ç–Ω–µ—Ä–≥–∏—è –≤–∞—à–µ–≥–æ –ø–∏—Ç–æ–º—Ü–∞ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç - {stats[2]}%, –∞ –≥–æ–ª–æ–¥ - {stats[0]}%')
            else:
                print('–ü–æ–∫–æ—Ä–º–∏—Ç–µ –≤–∞—à–µ —á–∞–¥–æ, –ø—Ä–µ–∂–¥–µ —á–µ–º –∏–¥—Ç–∏ —Å–ø–∞—Ç—å')
            print('-----------------------')
        elif question == "4":
            print(f"–ì–æ–ª–æ–¥ - {stats[0]}%, –†–∞–¥–æ—Å—Ç—å - {stats[1]}%, –≠–Ω–µ—Ä–≥–∏—è - {stats[2]}%")
            print('-----------------------')
        elif question == "5":
            break
    else:
        print('–í—ã –∫—Ä—É—Ç–æ–π, –≤–∞—à –ø–∏—Ç–æ–º–µ—Ü –∏–º–µ–µ—Ç —Ñ—É–ª–ª —Å—Ç–∞—Ç—ã!')


print(answers())

tasks()
````

# –ó–∞–¥–∞–Ω–∏–µ 2

–†—ã—Ü–∞—Ä—å –∏ –¥—Ä–∞–∫–æ–Ω

–¶–µ–ª—å: —Å–æ–∑–¥–∞—Ç—å –Ω–µ–±–æ–ª—å—à—É—é –∏–≥—Ä—É, –≤ –∫–æ—Ç–æ—Ä–æ–π –≤–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –∏–≥—Ä–∞—Ç—å –∑–∞ —Ä—ã—Ü–∞—Ä—è –∏ —Å—Ä–∞–∑–∏—Ç—å—Å—è —Å –¥—Ä–∞–∫–æ–Ω–æ–º

–¢—Ä–µ–±–æ–≤–∞–Ω–∏—è:

- –°–æ–∑–¥–∞–Ω–∏–µ –ø–µ—Ä—Å–æ–Ω–∞–∂–∞ (–∏–º—è, –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–æ—Å–ø–µ—Ö–∞—Ö, –æ—Ä—É–∂–∏–∏, —É—Ä–æ–Ω, –∑–¥–æ—Ä–æ–≤—å–µ)
- –£–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –ø–µ—Ä—Å–æ–Ω–∞–∂–µ–º –∏ –º–∏–Ω–∏ —Å—é–∂–µ—Ç
- –°–æ–∑–¥–∞–Ω–∏–µ –¥—Ä–∞–∫–æ–Ω–∞ (–ò–º—è, –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∑–¥–æ—Ä–æ–≤—å–µ –∏ —É—Ä–æ–Ω–µ)
- –ë–æ–µ–≤–∞—è —Å–∏—Å—Ç–µ–º–∞ (–Ω–∞–Ω–µ—Å–µ–Ω–∏–µ –∏ –ø–æ–ª—É—á–µ–Ω–∏–µ —É—Ä–æ–Ω–∞, —É—Ä–æ–Ω –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Å–ª—É—á–∞–π–Ω—ã–º –≤ –∑–∞–¥–∞–Ω–Ω–æ–º –¥–∏–∞–ø–∞–∑–æ–Ω–µ)
- –†–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –±–æ–π –º–µ–∂–¥—É –¥—Ä–∞–∫–æ–Ω–æ–º –∏ —Ä—ã—Ü–∞—Ä–µ–º

````python
import random

# –í —ç—Ç–æ—Ç —Ä–∞–∑ –±–µ–∑ –∞—Ä—Ç–µ–º–∏—è –ª–µ–±–µ–¥–µ–≤–∞ (

def create_character(character_type):
    name = input(f"–í–≤–µ–¥–∏—Ç–µ –∏–º—è –≤–∞—à–µ–≥–æ {character_type}: ")
    health = int(input(f"–í–≤–µ–¥–∏—Ç–µ –∑–¥–æ—Ä–æ–≤—å–µ –≤–∞—à–µ–≥–æ {character_type}: "))
    if character_type == "—Ä—ã—Ü–∞—Ä—è":
        weapon = input(f"–í—ã–±–µ—Ä–∏—Ç–µ –æ—Ä—É–∂–∏–µ –¥–ª—è –≤–∞—à–µ–≥–æ {character_type}: ")
    else:
        weapon = "–∞—Ä—Ç–µ–º–∏ –ª–µ–±–µ–¥–µ" # –¢—Å—Å—Å ü§´
    base_damage = int(input(f"–í–≤–µ–¥–∏—Ç–µ —É—Ä–æ–Ω –≤–∞—à–µ–≥–æ {character_type}: "))
    return {"name": name, "health": health, "weapon": weapon, "base_damage": base_damage}


def attack(attacker, defender):
    damage = random.randint(max(1, attacker["base_damage"] - 10), attacker["base_damage"] + 10)
    defender["health"] -= damage
    return damage


def block(character, damage):
    blocked_damage = damage // 2
    character["health"] -= blocked_damage
    return blocked_damage


def dragon_attack(dragon):
    if random.choice([True, False]):
        print("–î—Ä–∞–∫–æ–Ω –≥–æ—Ç–æ–≤–∏—Ç—Å—è –∫ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–º—É —É–¥–∞—Ä—É!")
        return random.randint(dragon["base_damage"] + 10, dragon["base_damage"] + 20)
    else:
        return random.randint(max(1, dragon["base_damage"] - 10), dragon["base_damage"] + 10)


print("–î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ –∏–≥—Ä—É!")
print(
    "–¢—ã - —Ä—ã—Ü–∞—Ä—å, –∫–æ—Ç–æ—Ä–æ–º—É –Ω–µ—á–µ–≥–æ –¥–µ–ª–∞—Ç—å, –ø–æ —ç—Ç–æ–º—É —Ç—ã —Ä–µ—à–∏–ª –Ω–∞–π—Ç–∏ —Å–µ–±–µ –¥—Ä–∞–∫–æ–Ω–∞ —á—Ç–æ–±—ã –ø–æ–¥—Ä–∞—Ç—å—Å—è —Å –Ω–∏–º")
print("–¢–µ–±–µ –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç —Å—Ä–∞–∑–∏—Ç—å—Å—è —Å –¥—Ä–∞–∫–æ–Ω–æ–º, –∫–æ—Ç–æ—Ä–æ–≥–æ —Ç—ã —Å–æ–∑–¥–∞—à—å —Å–∞–º. –£–¥–∞—á–∏ :)")

knight = create_character("—Ä—ã—Ü–∞—Ä—è")

dragon = create_character("–¥—Ä–∞–∫–æ–Ω–∞")

while knight["health"] > 0 and dragon["health"] > 0:
    print(f"\n{knight['name']} (–ó–¥–æ—Ä–æ–≤—å–µ: {knight['health']}) vs {dragon['name']} (–ó–¥–æ—Ä–æ–≤—å–µ: {dragon['health']})")

    dragon_damage = dragon_attack(dragon)

    action = input("–í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ (–∞—Ç–∞–∫–∞/–±–ª–æ–∫): ").lower()

    if action == "–∞—Ç–∞–∫–∞":
        damage = attack(knight, dragon)
        print(f"{knight['name']} –∞—Ç–∞–∫—É–µ—Ç {dragon['name']} –∏ –Ω–∞–Ω–æ—Å–∏—Ç {damage} —É—Ä–æ–Ω–∞")
    elif action == "–±–ª–æ–∫":
        blocked_damage = block(knight, dragon_damage)
        print(f"{knight['name']} –±–ª–æ–∫–∏—Ä—É–µ—Ç –∞—Ç–∞–∫—É –∏ –ø–æ–ª—É—á–∞–µ—Ç {blocked_damage} —É—Ä–æ–Ω–∞")
    else:
        print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤–≤–æ–¥")
        continue

    if dragon["health"] <= 0:
        print(f"{dragon['name']} –ø–æ–±–µ–∂–¥–µ–Ω, —Ç—ã –∫—Ä—É—Ç–æ–π")
        break

    if action != "–±–ª–æ–∫":
        knight["health"] -= dragon_damage
        print(f"{dragon['name']} –∞—Ç–∞–∫—É–µ—Ç –∏ –Ω–∞–Ω–æ—Å–∏—Ç {dragon_damage} —É—Ä–æ–Ω–∞ {knight['name']}")

    if knight["health"] <= 0:
        print(f"{knight['name']} –ø–æ–±–µ–∂–¥–µ–Ω, –∞–Ω–ª–∞–∫ :(")
        break
````

# –ó–∞–¥–∞–Ω–∏–µ 3


–¶–µ–ª—å - —Å–æ–∑–¥–∞—Ç—å –º–µ–Ω–µ–¥–∂–µ—Ä–∞ –∫–æ–º–∞–Ω–¥—ã Pok√©mon, –∫–æ—Ç–æ—Ä—ã–π –ø–æ–∑–≤–æ–ª–∏—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º:
