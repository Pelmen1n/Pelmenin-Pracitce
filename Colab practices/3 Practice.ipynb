Этап 1:
Детектор анаграмм Напишите программу на Python, которая принимает в качестве входных данных две строки и проверяет, являются ли они анаграммами друг друга

Этап 2:
Оформите ваш код в виде функции

Ввод:
```
listen, silent
```

Вывод:

```
True
```

````python
word1 = input()
word2 = input()
def anagram(word1, word2):
    list1 = list(word1)
    list2 = list(word2)
    list1.sort()
    list2.sort()
    position = 0
    matches = True
    while position < len(word1) and matches:
        if list1[position] == list2[position]:
            position = position + 1
        else:
            matches = False
    return matches
print(anagram('python', 'ythonp'))
````
